```python
# define a function that groups the data so that we have fewer age categories
# 0-4, 5-9, 10-17, 18-44, 45-64, 65+
# dropping row[0] as this includes characters in it, and so all values are converted to None
def rowmapper(row):
      return [row[0], row[1], row[2], row[3], row[4], row[5], row[6],
              row.HD01_VD05 + row.HD01_VD06, 
                row.HD01_VD07 + row.HD01_VD08 + row.HD01_VD09 + row.HD01_VD10 + row.HD01_VD11 + row.HD01_VD12 + row.HD01_VD13 + row.HD01_VD14,
                row.HD01_VD15 + row.HD01_VD16 + row.HD01_VD17 + row.HD01_VD18 + row.HD01_VD19,
                row.HD01_VD20 + row.HD01_VD21 + row.HD01_VD22 + row.HD01_VD23 + row.HD01_VD24 + row.HD01_VD25,
                row.HD01_VD26, row.HD01_VD27, row.HD01_VD28,
                row.HD01_VD29 + row.HD01_VD30, 
                row.HD01_VD31 + row.HD01_VD32 + row.HD01_VD33 + row.HD01_VD34 + row.HD01_VD35 + row.HD01_VD36 + row.HD01_VD37 + row.HD01_VD38,
                row.HD01_VD39 + row.HD01_VD40 + row.HD01_VD41 + row.HD01_VD42 + row.HD01_VD43,
                row.HD01_VD44 + row.HD01_VD45 + row.HD01_VD46 + row.HD01_VD47 + row.HD01_VD48 + row.HD01_VD49
                ]
```

```python
# create the table with a new header
table3 = etl.rowmap(table2, rowmapper, header=['GEO.id','GEO.id2', 'GEO.display-label', 'Estimate; Total', 'Estimate; Male'
                        ,'Male <5','Male 5-9', 'Male 10-17', 'Male 18-44', 'Male 45-64', 'Male 65+'
                        , 'Estimate: Female', 'Female <5','Female 5-9', 'Female 10-17', 
                                                'Female 18-44', 'Female 45-64', 'Female 65+'])
```
